This Notebook makes a spectral analysis of a sequence of energy resolution measurements.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import rfft

filename='/Users/laurent/Data/DUMPS/20191125_155318__0400_06_07_CAPOT_100/er_results.csv'
```

```python
def get_data(filename):
    import os
    if os.path.isfile(filename):
        print('OK! the file exists')
    f = open(filename, "r")
    data_txt=f.readlines()
    nresults=len(data_txt)-3
    data=np.zeros(nresults)
    for iline in range(nresults):
        data[iline]=float(data_txt[2+iline].split(';')[1])
    return(data)
```

```python
data_f=rfft(get_data(filename))
plt.plot(data_f[1:])
```
