```python
# Some tests with strings
 
str1 = "la tete a toto"

print(str1[0])
print(str1[-1])
# retournement d'un string par hachage
print(str1[len(str1)::-1])

print(str1[0::1])
print(str1[0::2])
```

```python
def palindrome(text):
    if text == text[len(text)::-1] :
        print(text, " is a palindrom")
        return(True)
    else :
        print(text, " is not a palindrom")
        return(False)
```

```python
a=['toto', 'otto', 'oto', 'kutaodufazuoayfgozfegup', 'azertyytreza']

for i in range(len(a)):
    palindrome(a[i])
```

```python
import timeit

text = 'azertyuiopoiuytreza'
timeit.timeit('palindrome(text)', number=1000, globals=globals())
```
