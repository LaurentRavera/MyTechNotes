```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
###############################################################################
def dec_to_signed_hexa(signeddec,nbits):
    r"""
    This function computes the hexadecimal string of a signed decimal value.
    
    Parameters
    ----------
    signeddec : decimal
        The decimal value to be converted
    nbits : number
        The number of bits of the data

    Output
    ------
    value : string
        The converted value

    Examples
    --------
    >>> dec_to_signed_hexa(2,8)                                                           
    '0x2'
    >>> dec_to_signed_hexa(2,16)                                                          
    '0x2'
    >>> dec_to_signed_hexa(-2,16)                                                         
    '0xfffe'
    """

    return(hex(signeddec & (2**nbits-1))[2:])
```

```python
def mk_sine_pattern(nval, fsr):
    pattern=np.round((fsr/2.)*np.sin(2.*np.pi*(1+np.arange(nval))/nval)+fsr/2).astype(int)

    return(pattern)
```

```python
def print_pattern(pattern):
    for i in range(len(pattern)):
        print('0x'+dec_to_signed_hexa(pattern[i],14))
```

```python
def print_pattern_to_file(filename, pattern):
    file = open(filename, 'w') #write to file
    for value in pattern:
        file.write('0x'+dec_to_signed_hexa(value,14)+'\n')
    file.close() #close file
```

```python
nval=680
fsr=2**14-1
filename='pattern_680_sine.txt'
pattern=mk_sine_pattern(nval, fsr)
plt.plot(pattern)
print_pattern(pattern)
print_pattern_to_file(filename, pattern)
```
