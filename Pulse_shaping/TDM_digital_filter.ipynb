> This Notebook simulates the digital filter at DAC output

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
npix=34
frow=5e6
ratio=20

fdac1=frow
fdac2=ratio*frow

pattern1=np.zeros(npix)
pattern2=np.zeros(npix*ratio)

for i in range(int(npix/2)):
    pattern1[2*i]=1
    for j in range(ratio):
        pattern2[2*i*ratio+j]=1

plt.plot(pattern2,'.')
```

```python
# Definition of the filter
def lpf1(vect):
    coeffs=np.array([
        -0.01575845273511803,
        -0.04130453806612838,
        -0.015379059726102796,
        0.10632777621080035,
        0.27697899483229904,
        0.3575220930099234,
        0.27697899483229904,
        0.10632777621080035,
        -0.015379059726102796,
        -0.04130453806612838,
        -0.01575845273511803])
    return(np.convolve(coeffs,vect,'same'))

def lpf2(vect):
    coeffs=np.array([
        -0.035947305873477074,
        0.01180074095982193,
        0.2890819817538472,
        0.48643738923365637,
        0.2890819817538472,
        0.01180074095982193,
        -0.035947305873477074])
    return(np.convolve(coeffs,vect,'same'))

def lpf3(vect):
    coeffs=np.array([
        -0.030253996832329176,
        0.276393202250021,
        0.5396029960012847,
        0.276393202250021,
        -0.030253996832329176])
    return(np.convolve(coeffs,vect,'same'))

def lpf4(vect):
    coeffs=np.array([
        -0.06852321822824987,
        0.2479193325683306,
        0.6399583674725382,
        0.2479193325683306,
        -0.06852321822824987])
    return(np.convolve(coeffs,vect,'same'))

def lpf5(vect):
    coeffs=signal.firwin(2, 0.01)
    return(np.convolve(coeffs,vect,'same'))
```

```python
fig=plt.figure(1,(12,8))
ax=fig.add_subplot(1,1,1)
ax.plot(pattern2, marker='.', label='Not filtered')
ax.plot(lpf5(pattern2), marker='.', label='Filtered')
ax.set_xlim(0,200)
fig.tight_layout()
```
