```python
import numpy as np
import matplotlib.pyplot as plt# Imports
```

```python
npts=2**12
input=(np.arange(npts)-npts/2)/npts
input_linear=input*0.3
input_limit=input*1.22
input_wide=input*3

fig=plt.figure(1,(14,8))
ax=fig.add_subplot(1,1,1)
ax.plot(input_wide, 2**11*np.sin(2*np.pi*input_wide), c='k', linewidth=1, label='SQUID transfer function')
ax.plot(input_limit, 2**11*np.sin(2*np.pi*input_limit), c='r', linewidth=4, label='Stable range')
ax.plot(input_linear, 2**11*np.sin(2*np.pi*input_linear), c='b', linewidth=6, label='Linear range')
ax.plot([-1.5,1.5], [2**11*0.95, 2**11*0.95], '--k', label='Anti delock triggers')
ax.plot([-1.5,1.5], [-2**11*0.95, -2**11*0.95], '--k')
ax.legend(loc='best')
ax.set_xlabel('SQUID input (Phi0)')
ax.set_ylabel('Firmware input (ADC code)')
for item in ([ax.title, ax.xaxis.label, ax.yaxis.label]):
    item.set_weight('bold')
for item in ([ax.title, ax.xaxis.label, ax.yaxis.label]):
    item.set_fontsize(17)
for item in (ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(15)

```
