> This notebook compares spectra obtained with different ki and kmix = 1.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
file_ki1p0_kmix0 = 'sc_spectra_ki1p0.npy'
file_ki1p2_kmix0 = 'sc_spectra_ki1p2.npy'
file_ki1p5_kmix0 = 'sc_spectra_ki1p5.npy'
file_ki2p0_kmix0 = 'sc_spectra_ki2p0.npy'
file_ki1p2_kmix1 = 'sc_spectra_ki1p2_kmix1.npy'
file_ki1p5_kmix1 = 'sc_spectra_ki1p5_kmix1.npy'
file_ki2p0_kmix1 = 'sc_spectra_ki2p0_kmix1.npy'
```

```python
def get_values(filename):
    with open(filename, 'rb') as file:
        freq = np.load(file)
        spectra = np.load(file)
    return(freq, spectra)
```

```python
"""
loading values with kmix=0
"""
freq, sp_ki1p0_kmix0 = get_values(file_ki1p0_kmix0)
freq, sp_ki1p2_kmix0 = get_values(file_ki1p2_kmix0)
freq, sp_ki1p5_kmix0 = get_values(file_ki1p5_kmix0)
freq, sp_ki2p0_kmix0 = get_values(file_ki2p0_kmix0)
```

```python
"""
Doing plots with kmix=0
"""
xlabel = r"Frequency (Hz)"
ylabel = r"Noise power spectral density vs baseline (dB/$\sqrt{Hz}$)"
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1)
ax.semilogx(freq[1:], sp_ki1p0_kmix0[0,1:], 'k', label='ki=1.0')
ax.semilogx(freq[1:], sp_ki1p2_kmix0[0,1:], 'r', label='ki=1.2')
ax.semilogx(freq[1:], sp_ki1p5_kmix0[0,1:], 'b', label='ki=1.5')
ax.semilogx(freq[1:], sp_ki2p0_kmix0[0,1:], 'g', label='ki=2.0')
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
for item in ([ax.title, ax.xaxis.label, ax.yaxis.label]):
    item.set_weight('bold')
    item.set_fontsize(14)
for item in (ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(12)
ax.legend(loc='lower left', fontsize=12)
ax.set_ylim([-110, -95])
ax.grid()
```

```python
"""
loading values with kmix=1
"""
freq, sp_ki1p2_kmix1 = get_values(file_ki1p2_kmix1)
freq, sp_ki1p5_kmix1 = get_values(file_ki1p5_kmix1)
freq, sp_ki2p0_kmix1 = get_values(file_ki2p0_kmix1)
```

```python
"""
Doing plots with kmix=1
"""
xlabel = r"Frequency (Hz)"
ylabel = r"Noise power spectral density vs baseline (dB/$\sqrt{Hz}$)"
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1)
ax.semilogx(freq[1:], sp_ki1p2_kmix1[0,1:], 'r', label='ki=1.2, kmix=1')
ax.semilogx(freq[1:], sp_ki1p5_kmix1[0,1:], 'g', label='ki=1.5, kmix=1')
ax.semilogx(freq[1:], sp_ki2p0_kmix1[0,1:], 'b', label='ki=2.0, kmix=1')
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
for item in ([ax.title, ax.xaxis.label, ax.yaxis.label]):
    item.set_weight('bold')
    item.set_fontsize(14)
for item in (ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(12)
ax.legend(loc='lower left', fontsize=12)
ax.set_ylim([-110, -95])
ax.grid()
```
